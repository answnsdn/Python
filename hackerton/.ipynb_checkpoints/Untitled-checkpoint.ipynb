```python
df1 = pd.read_csv('./rent_address.csv', encoding='cp949')
df1 = df1.dropna()
df1['대여소ID'].apply(int)
df_address = df1.loc[:,['대여소_구','대여소ID','대여소명','대여소주소','거치대수']]
df_address = df_address.sort_values(by='대여소ID', ascending=True)
import re
regex = re.compile('^\d+\.')
df_address['대여소명'] = df_address['대여소명'].apply(lambda s: regex.split(s)[-1].strip())
id = df_address['대여소ID'].astype('int')
df_address['대여소ID']=id
import numpy as np
address = np.where(df_address['대여소주소'].str.startswith(pat='서울특별시'),df_address['대여소주소'].str.split("서울특별시 ").str[1],df_address['대여소주소'])
df_address['대여소주소'] = address
df_address['대여소주소'] = '서울특별시 '+df_address['대여소주소']
df_address.loc[77,'대여소주소'] = '서울특별시 마포구 망원동'
df_address.to_csv('df_address.csv', encoding='cp949')
```

```python
#lambda라는 익명의 함수 사용하기
def station_name(s):
    return regex.split(s)[-1].strip()
df_address['대여소명'] = df_address['대여소명'].apply(station_name)

df_address['대여소명'] = df_address['대여소명'].apply(lambda s: regex.split(s)[-1].strip())
```

```python
rentals = pd.read_csv('./df_address.csv', encoding='cp949')
import re
regex = re.compile('(\d|-)+$')
rentals['대여소_동'] = rentals['대여소주소'].apply(lambda x: re.sub(regex, '', x.split()[2]))
rentals.info()
df = pd.DataFrame(rentals)
cols=['대여소_구','대여소_동','대여소명','대여소ID', '거치대수']
addlist = df.groupby(['대여소_구','대여소_동','대여소명','대여소ID','거치대수'],as_index=True)[cols].agg(lambda x: list(x)).reset_index(name="new")
del addlist['new']
addlist.rename(columns={"대여소_구":"rentGu"}, inplace=True)
addlist.rename(columns={"대여소_동":"rentDong"}, inplace=True)
addlist.rename(columns={"대여소명":"stationname"}, inplace=True)
addlist.rename(columns={"대여소ID":"stationnum"}, inplace=True)
addlist.rename(columns={"거치대수":"stationvalue"}, inplace=True)
addlist
#addlist.to_csv('addlist.csv')
# temp = addlist[addlist['대여소_구'].isin(['마포구'])]
# print(temp)
# dong = temp[temp['대여소_동']=='개포동']
# print(dong)
```
